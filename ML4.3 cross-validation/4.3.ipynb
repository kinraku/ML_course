#### Тестовая выборка

Для более правильного оценивания эффективности работы моделей разобьем исходную выборку на две части: тренировочную и тестовую:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)
```

В данном случае мы резервируем 15% данных для тестовой выборки.

Обучим модель логистической регрессии и выведем значения метрик:

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(solver='lbfgs')
model.fit(X_train, y_train) #Обучение трейновой выборке
y_pred = model.predict(X_test) #Предсказание для тестовой выборки
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
```

Видим, что модель обучилась до уровня точности 91%. Насколько это хорошо, судить сложно, надо сравнивать с результативностью других моделей. Для более наглядного представления результата можно вывести отчет о классификации:

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

Но заметим, что эта оценка производилась именно при данном разбиении. Что будет, если мы сделаем другое разбиение датасета на две части? Давайте повторим разбиение с другим значением random_state, обучим другую модель и выведем те же метрики:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)
model = LogisticRegression()
model.fit(X_train, y_train) #Обучение трейновой выборке
y_pred = model.predict(X_test) #Предсказание для тестовой выборки
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
```

Теперь получается, что модель обучилась идеально. Это же подтверждает и матрица:

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

#### Перекрестная проверка

Оценим работу построенной модели с помощью перекрёстной проверки.

Импортируем нужные библиотеки:

```python
from sklearn.model_selection import KFold,StratifiedKFold,LeaveOneOut, cross_val_score
```

k-fold разбиение
В k-блочной перекрёстной проверке исходные данные разбиваются на 
k (примерно) равных по количеству частей, называемых "блоками", на 𝑘−1 из которых производится обучение, а на 1 валидация. В результате получается более робастная оценка эффективности выбранной модели.

Создаём k-блочное разбиение (KFold):

```python
kf = KFold(n_splits = 3,shuffle=True, random_state=15)
kf
```

Метод split() - возвращает индексы разбиения:

Сделаем разбиение на блоки:

```python
for i, (train_index, test_index) in enumerate(kf.split(y)):
    print("Fold {}: Длинна train: {}, Длинна test: {}".format(i+1, len(train_index), len(test_index)))
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

Сделаем кросс-валидацию:

```python
metrics_accuracy = []
metrics_f1 = []
model = LogisticRegression(solver='lbfgs')
for i, (train_index, test_index) in enumerate(kf.split(y)):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    metrics_accuracy.append(accuracy_score(y_test, y_pred))
    metrics_f1.append(f1_score(y_test, y_pred, average='macro'))
```

Выведем соответствующие массивы с метриками:

```python
print('Значения метрики accuracy: {} \nЗначения метрики f1: {}'.format(metrics_accuracy, metrics_f1))
```

Выведем среднее значение метрики:

```python
print("Среднее по кросс-валидации: ", np.array(metrics_f1).mean())
```

Это и будет наша кросс-валидированная оценка метрики. Она гораздо ближе к истинному уровню эффективности модели за счет того, что все случайные ошибки выборки усредняются. Можно еще вывести дисперсию данной оценки, которая показывает степень уверенности в ней:

```python
import numpy as np
print("Среднее по кросс-валидации: ", np.array(metrics_f1).mean())
```

Выполняем кросс-валидацию с помощью функции cross_val_score:

```python
cv_results = cross_val_score(model,                  # модель
                             X,                      # матрица признаков
                             y,                      # вектор цели
                             cv = kf,                # тип разбиения (можно указать просто число фолдов cv = 3)
                             scoring = 'accuracy',   # метрика
                             n_jobs=-1)              # используются все ядра CPU

print("Кросс-валидация: ", cv_results)
print("Среднее по кросс-валидации: ", cv_results.mean())
print("Дисперсия по кросс-валидации: ", cv_results.std())
```

Этот код делает примерно то же, что и предыдущий, но автоматически, с использованием библиотечной функции.

Валидация по К-блокам (фолдам) - это самый распространенный алгоритм разбиения датасета на блоки. Он является золотым стандартом в научных исследованиях.

##### Stratified k-Fold

Метод stratified k-Fold — это метод k-Fold, использующий стратификацию при разбиении на фолды: каждый фолд содержит примерно такое же соотношение классов, как и всё исходное множество.
Такой подход может потребоваться в случае, например, очень несбалансированного соотношения классов.

Создаём стратифицированное k-блочное разбиение (StratifiedKFold):

```python
skf = StratifiedKFold(n_splits=3,shuffle=True, random_state=15)
skf.get_n_splits(X, y)
```

Выведем разбиение на блоки:

```python
for i, (train_index, test_index) in enumerate(skf.split(X,y)):
    print(f"Fold {i+1}:")
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

Выполняем кросс-валидацию с помощью функции cross_val_score:

```python
cv_results = cross_val_score(model,                  # модель
                             X,                      # матрица признаков
                             y,                      # вектор цели
                             cv = skf,           # тип разбиения
                             scoring = 'f1_macro',   # метрика
                             n_jobs=-1)              # используются все ядра CPU

print("Кросс-валидация: ", cv_results)
print("Среднее по кросс-валидации: ", cv_results.mean())
```

##### Leave-one-out

Метод leave-one-out (LOO) является частным случаем метода k-Fold: в нём каждый фолд состоит ровно из одного семпла.

Создаём разбиение:

```python
loo = LeaveOneOut()
```

Сделаем разбиение на блоки:

```python
for i, (train_index, test_index) in enumerate(loo.split(X)):
    print(f"Fold {i+1}:")
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

### Контрольные вопросы

**1. Почему для анализа модели нужно применять несколько метрик эффективности?**

Одна метрика, например accuracy, может быть неинформативна при несбалансированных классах: модель, предсказывающая всегда один класс, покажет высокую точность, но будет бесполезна. Precision, recall, F1 и другие метрики позволяют увидеть ошибки по каждому классу отдельно, оценить баланс между ложноположительными и ложноотрицательными срабатываниями и выбрать модель, подходящую под конкретную бизнес-задачу.

**2. Зачем для анализа качества модели делить датасет на обучающую и тренировочную выборки?**

Оценка качества на тех же данных, на которых модель обучалась, даёт завышенный (оптимистичный) результат, так как модель могла просто запомнить ответы (переобучение). Разделение позволяет проверить обобщающую способность – способность модели правильно предсказывать на новых, ранее не виденных данных. Это единственный способ оценить, как модель будет работать в реальных условиях.

**3. В чем особенность и область применения каждой метрики качества?**

- **Accuracy** – доля правильных ответов. Хороша при сбалансированных классах.
- **Precision** – из всех предсказанных положительных – сколько действительно положительных. Важна, когда ложноположительные срабатывания дороги (спам-фильтр).
- **Recall** – из всех реальных положительных – сколько модель нашла. Важна, когда ложноотрицательные срабатывания опасны (диагностика рака).
- **F1-мера** – среднее гармоническое precision и recall. Используется, когда нужен баланс между ними.
- **Confusion matrix** – показывает все четыре исхода (TP, TN, FP, FN) и позволяет детально разобрать ошибки по каждому классу.

**4. Для каждой метрики классификации и регрессии придумайте пример, в котором данная метрика будет определяющей для выбора наилучшей модели.**

- **Accuracy** – задача угадывания пола по росту и весу (классы примерно поровну).
- **Precision** – поисковая выдача: важно, чтобы на первой странице были только релевантные результаты.
- **Recall** – детекция террористических угроз: лучше перестраховаться и проверить лишнего, чем пропустить реальную угрозу.
- **F1** – чат-бот: важно и отвечать по делу (precision), и не пропускать вопросы (recall).
- **MAE (регрессия)** – прогноз температуры: проще интерпретировать в градусах.
- **MSE/RMSE** – штрафует большие ошибки сильнее: подходит, когда крупные отклонения недопустимы (управление ракетой).
- **R²** – сравнение качества моделей на разных шкалах целевой переменной.

**5. В каких случаях нельзя делить выборку случайным образом?**

- Данные имеют временную структуру (временные ряды, прогнозирование курса акций) – тогда нужно делить по времени: обучающая выборка – прошлое, тестовая – будущее, иначе произойдёт «подглядывание в будущее» (data leakage).
- Данные сгруппированы (несколько образцов от одного пациента) – тогда нужно делить по группам, чтобы один пациент не попал и в обучение, и в тест.
- Данные имеют пространственную автокорреляцию (например, спутниковые снимки соседних участков) – требуется специальное пространственное разделение.

**6. Зачем нужен и как использовать отчет о классификации?**

Отчёт о классификации (classification_report) выдаёт для каждого класса значения precision, recall, f1-score и support (количество реальных примеров класса). Он позволяет быстро оценить, на каких классах модель работает хорошо, а на каких плохо, увидеть дисбаланс метрик и принять решение о необходимости сбора данных по плохо распознаваемому классу или выборе другой модели. Используется после обучения, на тестовой выборке, передавая в функцию истинные и предсказанные метки.

**7. В чем разница между KFold и StratifiedKFold?**

KFold разбивает данные на k блоков случайным образом, не учитывая распределение классов. StratifiedKFold также разбивает на k блоков, но сохраняет пропорцию классов в каждом фолде такой же, как в исходном датасете. Это особенно важно при сильном дисбалансе классов, чтобы в обучающей и валидационной выборках были представлены все классы.

**8. Когда целесообразно использовать LeaveOneOut?**

LeaveOneOut (LOO) целесообразно использовать, когда размер датасета очень маленький (десятки или сотни объектов). LOO даёт максимально возможное количество обучающих примеров для каждой итерации (n-1 из n) и минимальную смещённость оценки. Однако LOO вычислительно затратен (требует n итераций) и может иметь высокую дисперсию оценки качества.
